```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
pd.set_option('display.max_columns', 60)
```

```python
df = pd.read_csv('./spam_email.csv', sep='\s+', engine='python', header=0)
df.rename(columns={'#a':'a'}, inplace=True)
df.head()
```

```python
df.describe()
```

```python
df.isna().any().any()
```

```python
corr=df.corr()
plt.figure(figsize=(16, 12))
ax=sns.heatmap(corr,
            xticklabels=corr.columns,
            yticklabels=corr.columns,
            cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True,
            linewidths=.1)
ax.set(title="all variables correlation heatmap");
```

```python
k=10 #number of variables
cols=corr.abs().nlargest(k,'make')['make'].index
cm=np.corrcoef(df[cols].values.T)
plt.figure(figsize=(10, 8))
ax=sns.heatmap(cm,yticklabels=cols.values,
                xticklabels=cols.values,
                cmap=sns.diverging_palette(220, 10, as_cmap=True),
                annot=True,
                square=True)
ax.set(title=f"top {k} variables correlation with class heatmap");
plt.show()
```

```python
pivot_class_freq=pd.pivot_table(df,values=df.drop(["make","cap_avg","cap_long","cap_total"],axis=1),index='make',aggfunc="mean")
pivot_class_freq
```

```python
plt.figure(figsize=(16,8))
plt.xticks(rotation=90)
sns.barplot(x=pivot_class_freq.columns,
            y=pivot_class_freq.iloc[0]-pivot_class_freq.iloc[1]).set(title="Non Spam minus Spam average word and char frequency",ylabel="average frequency difference");
```

```python
pivot_class_cap=pd.pivot_table(df,values=df[["cap_avg","cap_long","cap_total"]],index='make',aggfunc="mean")
pivot_class_cap
```

```python
plt.xticks(rotation=30)
sns.barplot(x=pivot_class_cap.columns,
            y=pivot_class_cap.iloc[0]-pivot_class_cap.iloc[1]).set(title="Non Spam minus Spam",ylabel="average difference");
```
